## Question 5: Plot Tesla Stock Graph

Use the make_graph function to graph the Tesla Stock Data, also provide a title for the graph.

```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Step 1: Download Tesla stock data
tesla_data = yf.download('TSLA', start='2010-01-01', end='2023-01-01')
```

```python
# Step 2: Define the make_graph function
def make_graph(data, title):
    plt.figure(figsize=(3, 2))
    plt.plot(data.index, data['Close'], label='Tesla Stock Price')
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
# Step 3: Use the make_graph function to plot the Tesla stock data
make_graph(tesla_data, 'Tesla Stock Price Over Time')
```
