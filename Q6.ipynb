## Question 6: Plot GameStop Stock Graph

Use the make_graph function to graph the GameStop Stock Data, also provide a title for the graph.

```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Step 1: Download Gamestop stock data
gme_data = yf.download('GME', start='2010-01-01', end='2023-01-01')
```

```python
# Step 2: Define the make_graph function
def make_graph(data, title):
    plt.figure(figsize=(3, 2))
    plt.plot(data.index, data['Close'], label='Gamestop Stock Price')
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
# Step 3: Use the make_graph function to plot the Gamestop stock data
make_graph(gme_data, 'Gamestop Stock Price Over Time')
```
